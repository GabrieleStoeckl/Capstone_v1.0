## Remove intial purchase cycle
Since we are analysing a dataset with a finite start day, each product-by-household combination will have a first occurrence. However, this may just be the first event that shows up in our data but actually is part of an ongoing purchasing behaviour, and therefore no true first / trial purchase. 
Based on the average purchase frequency of products in our category, we can make an informed differentiation between recurring and new purchases.
The average purchase frequency is 7.9 weeks. We have cut off the first 12 weeks due to doubts about data completeness. If in addition we allow two full category purchase cycles of 16 weeks, cutting off the first 30 weeks should be a good approximation towards identifying real new product purchases.

```python
average_purchase_frequency(candy)
```

```python
remove_weeks_first_purchase_cycle(candy, 30)
```
