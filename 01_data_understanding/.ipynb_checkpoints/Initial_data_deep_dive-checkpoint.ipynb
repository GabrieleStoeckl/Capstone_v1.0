```python
first_occurence = pd.DataFrame(soft_drinks.groupby(["household_key", "PRODUCT_ID"])["WEEK_NO"].agg(["min", "max", "count"]))
```

```python
first_occurence.describe()
```

```python
first_occurence["count"].value_counts()
```

```python
first_occurence.rename(columns = {"max" : "last_week", "min" : "first_week", "count" :"no_weeks"}, inplace = True)
first_occurence.head()
```

```python
first_occurence.info()
```

```python
for index, row in first_occurence.iterrows():
    if row.no_weeks > 1:
        first_occurence.loc[index, "time_lag"] = row.last_week - row.first_week
        first_occurence.loc[index, "freq_weeks"] = first_occurence.loc[index, "time_lag"] / row.no_weeks
    else:
        continue
```
