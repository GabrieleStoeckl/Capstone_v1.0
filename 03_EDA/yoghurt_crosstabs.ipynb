# Inspect feature means by target variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("../02_data_preparation/yoghurt_w_dummy_data.csv")
data.info()
```

```python
data.drop(labels=["Unnamed: 0", "display_1", "display_5", "display_A", "campaign_6.0", "campaign_30.0", "description_TypeC"],
          axis = 1, inplace = True)
```

```python
pd.set_option("display.max_columns", None)
data.describe(include = "all")
```

```python
pd.crosstab(data["description_TypeA"], data["first_purchase"], normalize = "columns")
```

```python
new_data = pd.DataFrame(data.groupby(["first_purchase"]).agg(["mean"])).transpose()
new_data
```

```python
#add a columns with difference between means True vs False
new_data["DIFF"] = new_data[True] - new_data[False]
new_data
```
