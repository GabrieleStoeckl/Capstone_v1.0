```python
#Append category information to transactions
transactions = pd.merge(transactions, products, how = "left", on = "PRODUCT_ID")
transactions.info()
```

```python
transactions["COMMODITY_DESC"].value_counts()
```

```python
#add column with first purchase per product per household
transactions["first_purchase"] = 0
transactions["first_purchase"].loc[transactions.groupby(["household_key", "PRODUCT_ID"]).DAY.idxmin()] = 1
transactions["first_purchase"].value_counts()
```

```python
transactions["first_purchase"].value_counts()
```

```python
cheese = transactions[transactions["COMMODITY_DESC"].isin( 
                         ["CHEESE", "CHEESES"])]
cheese.info()
```

```python
cheese["COMMODITY_DESC"].value_counts()
```

```python
#calculate regular shelf price w/o discounts
cheese["shelf_price"] = (cheese["SALES_VALUE"] - cheese["RETAIL_DISC"] - cheese["COUPON_MATCH_DISC"]) / cheese["QUANTITY"]
#calculate paid shelf price with discounts
cheese["paid_price"] = (cheese["SALES_VALUE"] + cheese["COUPON_DISC"]) / cheese["QUANTITY"]
#calculate overall discount
cheese["pct_disc"] = (cheese["shelf_price"]-cheese["paid_price"]) / cheese["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to retailer loyalty card 
cheese["pct_retail_disc"] = (-cheese["RETAIL_DISC"] / cheese["QUANTITY"]) / cheese["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to manufacturer coupons 
cheese["pct_coupon_disc"] = ((-cheese["COUPON_DISC"]-cheese["COUPON_MATCH_DISC"]) / cheese["QUANTITY"] ) / cheese["shelf_price"]
```

```python
#adding in causal data
causal = pd.read_parquet("data/causal_data.parquet")
causal.info()
```

```python
causal_cheese = pd.merge(causal, products[["PRODUCT_ID", "COMMODITY_DESC"]], how = "left", on = "PRODUCT_ID")
```

```python
causal_cheese = causal_cheese[causal_cheese.COMMODITY_DESC.isin(
    ["CHEESE", "CHEESES"])]
causal_cheese.info()
```

```python
causal_cheese["COMMODITY_DESC"].value_counts()
```

```python
cheese = pd.merge(cheese, causal_cheese, how = "left", on = ["PRODUCT_ID", "STORE_ID", "WEEK_NO"])
cheese.info()
```

```python
#remove duplicates
cheese.drop_duplicates(subset= ["household_key", "BASKET_ID", "STORE_ID", "DAY", "PRODUCT_ID"], keep = "last", inplace= True)
cheese.info()
```

```python
#dummy variables for display
dummies_display = pd.get_dummies(cheese["display"], prefix = "display")
cheese = pd.concat((cheese, dummies_display), axis = 1)
#dummy variables for mailer
dummies_mailer = pd.get_dummies(cheese["mailer"], prefix = "mailer")
cheese = pd.concat((cheese, dummies_mailer), axis = 1)
```

```python
cheese.drop("COMMODITY_DESC_y", axis = 1, inplace = True)
cheese.rename(columns = {"COMMODITY_DESC_x" : "COMMODITY_DESC"}, inplace=True)
cheese.head()
```

```python
campaign_data = campaign_data[campaign_data["COUPON_UPC"] > 0]
campaign_data.info()
```

```python
cheese = cheese[cheese.age_ordinal > 0]
cheese.describe()
```

```python
#remove first 30 weeks
cheese.drop(cheese[cheese.WEEK_NO < 30].index, inplace = True)
cheese.describe()
```

```python
cheese_wcamp = pd.merge(cheese, campaign_data, how = "left", on = ["household_key", "PRODUCT_ID"])
cheese_wcamp.info()
```

```python
cheese_wcamp["CAMPAIGN_y"] = np.where((cheese_wcamp["START_DAY"] <= cheese_wcamp["DAY"]) & 
                                           (cheese_wcamp["DAY"] <= cheese_wcamp["END_DAY"]), 
                                           cheese_wcamp["CAMPAIGN_y"], np.NaN)
cheese_wcamp["DESCRIPTION"] = np.where((cheese_wcamp["START_DAY"] <= cheese_wcamp["DAY"]) & 
                                           (cheese_wcamp["DAY"] <= cheese_wcamp["END_DAY"]), 
                                           cheese_wcamp["DESCRIPTION"], np.NaN)
```

```python
cheese_wcamp["DESCRIPTION"] = np.where((cheese_wcamp["START_DAY"] <= cheese_wcamp["DAY"]) & 
                                           (cheese_wcamp["DAY"] <= cheese_wcamp["END_DAY"]), 
                                           cheese_wcamp["DESCRIPTION"], np.NaN)
```

```python
cheese_wcamp.drop(labels = ["MANUFACTURER_y", "DEPARTMENT_y", "BRAND_y", "COMMODITY_DESC_y", "SUB_COMMODITY_DESC_y", 
                                 "CURR_SIZE_OF_PRODUCT_y"], 
                       axis = 1, inplace = True)
```

```python
cheese_wcamp.info()
```

```python
cheese_wcamp.sort_values(by = ["household_key", "PRODUCT_ID", "BASKET_ID", "CAMPAIGN_y"], inplace = True)
cheese_wcamp.drop_duplicates(subset = ["household_key", "PRODUCT_ID", "BASKET_ID"], keep = "first", inplace = True)
cheese_wcamp.info()
```

```python
cheese_wcamp.to_csv("data/cheese_wcamp.csv")
```
