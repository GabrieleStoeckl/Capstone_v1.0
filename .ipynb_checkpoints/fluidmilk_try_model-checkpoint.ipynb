```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fluid_milk = pd.read_csv("data/fluid_milk_wcamp.csv")
fluid_milk.info()
```

```python
fluid_milk.drop(labels = ["Unnamed: 0", "household_key", "BASKET_ID", "PRODUCT_ID", "QUANTITY", "STORE_ID", 
                           "RETAIL_DISC", "TRANS_TIME", "COUPON_DISC", "COUPON_MATCH_DISC", "MANUFACTURER_x", 
                           "DEPARTMENT_x", "BRAND_x", "COMMODITY_DESC_x", "SUB_COMMODITY_DESC_x", "CURR_SIZE_OF_PRODUCT_x", 
                           "display", "mailer", "display_0", "AGE_DESC", "MARITAL_STATUS_CODE", "INCOME_DESC", 
                           "HOMEOWNER_DESC", "HH_COMP_DESC", "HOUSEHOLD_SIZE_DESC", "KID_CATEGORY_DESC",
                           "COUPON_UPC_y", "START_DAY", "END_DAY", "CAMPAIGN_y", "DESCRIPTION"], 
                 axis = 1, inplace = True)
pd.set_option("display.max_columns", None)
fluid_milk.head()
```

```python
#remove first 30 weeks
fluid_milk.drop(fluid_milk[fluid_milk.WEEK_NO < 30].index, inplace = True)
fluid_milk.describe()
```

```python
fluid_milk.drop(labels = ["DAY", "WEEK_NO", "mailer_0"], axis = 1, inplace = True)
fluid_milk.describe()
```

```python
fluid_milk.drop(labels = ["COUPON_UPC_x", "CAMPAIGN_x", "SALES_VALUE"], axis = 1, inplace = True)
fluid_milk.describe()
```

```python
fluid_milk.tail()
```

```python
fluid_milk["first_purchase"] = fluid_milk["first_purchase"].fillna(0)
fluid_milk.describe()
```

```python
fluid_milk["age_ordinal"] = fluid_milk["age_ordinal"].fillna(0)
fluid_milk.describe()
```

```python
#remove samples without HH demogs
fluid_milk.drop(fluid_milk[fluid_milk.age_ordinal <1].index, inplace = True)
fluid_milk.describe()
```

```python
from pandas_profiling import ProfileReport
```

```python
prof = ProfileReport(fluid_milk)
prof.to_file(output_file='data/output_fluid_milk.html')
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from sklearn.model_selection import learning_curve, validation_curve

from sklearn.metrics import f1_score
```

```python
X = fluid_milk.drop("first_purchase", axis = 1)
y = fluid_milk["first_purchase"]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, stratify = y)
```

```python
clf = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits = 5) 
scores = cross_val_score(clf, X_train, y_train, cv=cv, scoring = "f1" )
print(scores)
```
