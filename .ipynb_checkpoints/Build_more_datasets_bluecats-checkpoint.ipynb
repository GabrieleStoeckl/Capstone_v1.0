```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
products.info()
```

```python
#Append category information to transactions
transactions = pd.merge(transactions, products, how = "left", on = "PRODUCT_ID")
transactions.info()
```

```python
transactions.head()
```

```python
transactions["COMMODITY_DESC"].value_counts()
```

```python
#add column with first purchase per product per household
transactions["first_purchase"] = 0
transactions["first_purchase"].loc[transactions.groupby(["household_key", "PRODUCT_ID"]).DAY.idxmin()] = 1
transactions["first_purchase"].value_counts()
```

```python
transactions["first_purchase"].value_counts()
```

```python
blue_cats = transactions[transactions["COMMODITY_DESC"].isin( 
                         ["DEODORANTS", "DISHWASH DETERGENTS", "FEMININE HYGIENE", "HAIR CARE PRODUCTS", "HOUSEHOLD CLEANG NEEDS", "LAUNDRY ADDITIVES",
                         "LAUNDRY DETERGENTS", "ORAL HYGIENE PRODUCTS", "SHAVING CARE PRODUCTS"])]
blue_cats.info()
```

```python
blue_cats["COMMODITY_DESC"].value_counts()
```

```python
#calculate regular shelf price w/o discounts
blue_cats["shelf_price"] = (blue_cats["SALES_VALUE"] - blue_cats["RETAIL_DISC"] - blue_cats["COUPON_MATCH_DISC"]) / blue_cats["QUANTITY"]
#calculate paid shelf price with discounts
blue_cats["paid_price"] = (blue_cats["SALES_VALUE"] + blue_cats["COUPON_DISC"]) / blue_cats["QUANTITY"]
#calculate overall discount
blue_cats["pct_disc"] = (blue_cats["shelf_price"]-blue_cats["paid_price"]) / blue_cats["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to retailer loyalty card 
blue_cats["pct_retail_disc"] = (-blue_cats["RETAIL_DISC"] / blue_cats["QUANTITY"]) / blue_cats["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to manufacturer coupons 
blue_cats["pct_coupon_disc"] = ((-blue_cats["COUPON_DISC"]-blue_cats["COUPON_MATCH_DISC"]) / blue_cats["QUANTITY"] ) / blue_cats["shelf_price"]
```

```python
#adding in causal data
causal = pd.read_parquet("data/causal_data.parquet")
causal.info()
```

```python
causal_blue_cats = pd.merge(causal, products[["PRODUCT_ID", "COMMODITY_DESC"]], how = "left", on = "PRODUCT_ID")
```

```python
causal_blue_cats = causal_blue_cats[causal_blue_cats.COMMODITY_DESC.isin(
    ["DEODORANTS", "DISHWASH DETERGENTS", "FEMININE HYGIENE", "HAIR CARE PRODUCTS", "HOUSEHOLD CLEANG NEEDS", "LAUNDRY ADDITIVES", 
     "LAUNDRY DETERGENTS", "ORAL HYGIENE PRODUCTS", "SHAVING CARE PRODUCTS"])]
causal_blue_cats.info()
```

```python
blue_cats = pd.merge(blue_cats, causal_blue_cats, how = "left", on = ["PRODUCT_ID", "STORE_ID", "WEEK_NO"])
blue_cats.info()
```

```python
#remove duplicates
blue_cats.drop_duplicates(subset= ["household_key", "BASKET_ID", "STORE_ID", "DAY", "PRODUCT_ID"], keep = "last", inplace= True)
blue_cats.info()
```

```python
#dummy variables for display
dummies_display = pd.get_dummies(blue_cats["display"], prefix = "display")
blue_cats = pd.concat((blue_cats, dummies_display), axis = 1)
#dummy variables for mailer
dummies_mailer = pd.get_dummies(blue_cats["mailer"], prefix = "mailer")
blue_cats = pd.concat((blue_cats, dummies_mailer), axis = 1)
```

```python
blue_cats.drop("COMMODITY_DESC_y", axis = 1, inplace = True)
blue_cats.rename(columns = {"COMMODITY_DESC_x" : "COMMODITY_DESC"}, inplace=True)
blue_cats.head()
```

```python
blue_cats = blue_cats[blue_cats.age_ordinal > 0]
blue_cats.describe()
```

```python
#remove first 30 weeks
blue_cats.drop(blue_cats[blue_cats.WEEK_NO < 30].index, inplace = True)
blue_cats.describe()
```

```python
blue_cats_wcamp = pd.merge(blue_cats, campaign_data, how = "left", on = ["household_key", "PRODUCT_ID"])
blue_cats_wcamp.info()
```

```python
blue_cats_wcamp["CAMPAIGN_y"] = np.where((blue_cats_wcamp["START_DAY"] <= blue_cats_wcamp["DAY"]) & 
                                           (blue_cats_wcamp["DAY"] <= blue_cats_wcamp["END_DAY"]), 
                                           blue_cats_wcamp["CAMPAIGN_y"], np.NaN)
blue_cats_wcamp["DESCRIPTION"] = np.where((blue_cats_wcamp["START_DAY"] <= blue_cats_wcamp["DAY"]) & 
                                           (blue_cats_wcamp["DAY"] <= blue_cats_wcamp["END_DAY"]), 
                                           blue_cats_wcamp["DESCRIPTION"], np.NaN)
```

```python
blue_cats_wcamp["DESCRIPTION"] = np.where((blue_cats_wcamp["START_DAY"] <= blue_cats_wcamp["DAY"]) & 
                                           (blue_cats_wcamp["DAY"] <= blue_cats_wcamp["END_DAY"]), 
                                           blue_cats_wcamp["DESCRIPTION"], np.NaN)
```

```python
blue_cats_wcamp.drop(labels = ["MANUFACTURER_y", "DEPARTMENT_y", "BRAND_y", "COMMODITY_DESC_y", "SUB_COMMODITY_DESC_y", 
                                 "CURR_SIZE_OF_PRODUCT_y"], 
                       axis = 1, inplace = True)
```

```python
blue_cats_wcamp.info()
```

```python
blue_cats_wcamp.sort_values(by = ["household_key", "PRODUCT_ID", "BASKET_ID", "CAMPAIGN_y"], inplace = True)
blue_cats_wcamp.drop_duplicates(subset = ["household_key", "PRODUCT_ID", "BASKET_ID"], keep = "first", inplace = True)
blue_cats_wcamp.info()
```

```python
blue_cats_wcamp.to_csv("data/blue_cats_wcamp.csv")
```
