```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
candy = pd.read_csv("data/candy_wcamp.csv")
candy.info()
```

```python
candy.drop(labels = ["Unnamed: 0", "Unnamed: 0.1", "household_key", "BASKET_ID", "PRODUCT_ID", "QUANTITY", "STORE_ID", 
                           "RETAIL_DISC", "TRANS_TIME", "COUPON_DISC", "COUPON_MATCH_DISC", "MANUFACTURER_x", 
                           "DEPARTMENT_x", "BRAND_x", "COMMODITY_DESC_x", "SUB_COMMODITY_DESC_x", "CURR_SIZE_OF_PRODUCT_x", 
                           "display", "mailer", "display_0", "AGE_DESC", "MARITAL_STATUS_CODE", "INCOME_DESC", 
                           "HOMEOWNER_DESC", "HH_COMP_DESC", "HOUSEHOLD_SIZE_DESC", "KID_CATEGORY_DESC",
                           "COUPON_UPC_y", "START_DAY", "END_DAY", "CAMPAIGN_y", "DESCRIPTION"], 
                 axis = 1, inplace = True)
pd.set_option("display.max_columns", None)
candy.head()
```

```python
#remove first 13 weeks
candy.drop(candy[candy.WEEK_NO < 13].index, inplace = True)
candy.describe()
```

```python
candy.drop(labels = ["DAY", "WEEK_NO", "mailer_0"], axis = 1, inplace = True)
candy.describe()
```

```python
candy.drop(labels = ["COUPON_UPC_x", "CAMPAIGN_x", "SALES_VALUE"], axis = 1, inplace = True)
candy.describe()
```

```python
candy.tail()
```

```python
candy["first_purchase"] = candy["first_purchase"].fillna(0)
candy.describe()
```

```python
candy["age_ordinal"] = candy["age_ordinal"].fillna(0)
candy.describe()
```

```python
#remove samples without HH demogs
candy.drop(candy[candy.age_ordinal <1].index, inplace = True)
candy.describe()
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from sklearn.model_selection import learning_curve, validation_curve

from sklearn.metrics import f1_score, roc_auc_score
```

```python
X = candy.drop("first_purchase", axis = 1)
y = candy["first_purchase"]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, stratify = y)
```

```python
clf = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits = 5) 
scores = cross_val_score(clf, X_train, y_train, cv=cv, scoring = "f1" )
print(scores)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.feature_importances_
```

```python
importances = clf.feature_importances_

indices = np.argsort(importances)[::-1]

# rearrange feature names so they match the sorted feature names
# I have my original names list of feature names, which corresponds to the features in the tree.
f_names = [X.columns [i] for i in indices]

#create my plot
fig, axes = plt.subplots(figsize=(20,5))
axes.set_title("Feature Importances")
axes.bar(range(X_train.shape[1]), importances[indices])

# Add feature names as x-axis labels
axes.set_xticks(range(X.shape[1]));
axes.set_xticklabels(f_names, rotation=45, ha='right');
```

```python
import pickle
```

```python
filename = "candy_dumb_clf_model.pck"
pickle.dump(clf, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```

```python
y_pred = loaded_model.predict(X_test)
```
