```python
import numpy as np
import pandas as pd
```

```python
campaign_desc = pd.read_csv("data/campaign_desc.csv")
coupon = pd.read_csv("data/coupon.csv")
campaign_table = pd.read_csv("data/campaign_table.csv")
products = pd.read_csv("data/product.csv")
```

```python
products = products[products["COMMODITY_DESC"] == "SOFT DRINKS"]
products.info()
```

```python
campaign_data = pd.merge(products, coupon, how = "left", on = "PRODUCT_ID")
campaign_data.info()
```

```python
campaign_data = pd.merge(campaign_data, campaign_desc, how = "left", on = "CAMPAIGN")
campaign_data.info()
```

```python
campaign_data.head()
```

```python
campaign_data = pd.merge(campaign_data, campaign_table, how = "left", on = "CAMPAIGN")
campaign_data.info()
```

```python
campaign_data.drop("DESCRIPTION_y", axis = 1, inplace = True)
```

```python
campaign_data.rename(columns = {"DESCRIPTION_x" : "DESCRIPTION"}, inplace = True)
```

```python
campaign_data = campaign_data[campaign_data["COUPON_UPC"] > 0]
campaign_data.info()
```

```python
soft_drinks_wcamp.sort_values(by = ["household_key", "PRODUCT_ID", "BASKET_ID", "CAMPAIGN_y"], inplace = True)
soft_drinks_wcamp.drop_duplicates(subset = ["household_key", "PRODUCT_ID", "BASKET_ID"], keep = "first", inplace = True)
soft_drinks_wcamp.info()
```

```python
soft_drinks_wcamp.to_csv("data/soft_drinks_wcamp.csv")
```

```python
#dummify campaign and description of campaign
soft_drinks_wcamp = pd.read_csv("data/soft_drinks_wcamp.csv")
dummies_campaign_y = pd.get_dummies(soft_drinks_wcamp["CAMPAIGN_y"], prefix = "campaign_y")
dummies_description = pd.get_dummies(soft_drinks_wcamp["DESCRIPTION"], prefix = "description")
soft_drinks_wcamp = pd.concat((soft_drinks_wcamp, dummies_campaign_y), axis = 1)
soft_drinks_wcamp = pd.concat((soft_drinks_wcamp, dummies_description), axis = 1)
soft_drinks_wcamp.info()
```

```python
soft_drinks_wcamp.to_csv("data/soft_drinks_wcamp.csv")
```
