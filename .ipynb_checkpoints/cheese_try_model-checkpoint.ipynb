```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
cheese = pd.read_csv("data/cheese_wcamp.csv")
cheese.info()
```

```python
cheese.drop(labels = ["Unnamed: 0", "Unnamed: 0.1", "household_key", "BASKET_ID", "PRODUCT_ID", "QUANTITY", "STORE_ID", 
                           "RETAIL_DISC", "TRANS_TIME", "COUPON_DISC", "COUPON_MATCH_DISC", "MANUFACTURER_x", 
                           "DEPARTMENT_x", "BRAND_x", "COMMODITY_DESC_x", "SUB_COMMODITY_DESC_x", "CURR_SIZE_OF_PRODUCT_x", 
                           "display", "mailer", "display_0", "AGE_DESC", "MARITAL_STATUS_CODE", "INCOME_DESC", 
                           "HOMEOWNER_DESC", "HH_COMP_DESC", "HOUSEHOLD_SIZE_DESC", "KID_CATEGORY_DESC",
                           "COUPON_UPC_y", "START_DAY", "END_DAY", "CAMPAIGN_y", "DESCRIPTION"], 
                 axis = 1, inplace = True)
pd.set_option("display.max_columns", None)
cheese.head()
```

```python
#remove first 13 weeks
cheese.drop(cheese[cheese.WEEK_NO < 13].index, inplace = True)
cheese.describe()
```

```python
cheese.drop(labels = ["DAY", "WEEK_NO", "mailer_0"], axis = 1, inplace = True)
cheese.describe()
```

```python
cheese.drop(labels = ["COUPON_UPC_x", "CAMPAIGN_x", "SALES_VALUE"], axis = 1, inplace = True)
cheese.describe()
```

```python
cheese.tail()
```

```python
cheese["first_purchase"] = cheese["first_purchase"].fillna(0)
cheese.describe()
```

```python
cheese["age_ordinal"] = cheese["age_ordinal"].fillna(0)
cheese.describe()
```

```python
#remove samples without HH demogs
cheese.drop(cheese[cheese.age_ordinal <1].index, inplace = True)
cheese.describe()
```

```python
from pandas_profiling import ProfileReport
```

```python
prof = ProfileReport(cheese)
prof.to_file(output_file='data/output_cheese.html')
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit, StratifiedKFold
from sklearn.model_selection import learning_curve, validation_curve

from sklearn.metrics import f1_score
```

```python
X = cheese.drop("first_purchase", axis = 1)
y = cheese["first_purchase"]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, stratify = y)
```

```python
clf = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits = 5) 
scores = cross_val_score(clf, X_train, y_train, cv=cv, scoring = "f1" )
print(scores)
```
