```python
#Append category information to transactions
transactions = pd.merge(transactions, products, how = "left", on = "PRODUCT_ID")
transactions.info()
```

```python
transactions["COMMODITY_DESC"].value_counts()
```

```python
#add column with first purchase per product per household
transactions["first_purchase"] = 0
transactions["first_purchase"].loc[transactions.groupby(["household_key", "PRODUCT_ID"]).DAY.idxmin()] = 1
transactions["first_purchase"].value_counts()
```

```python
transactions["first_purchase"].value_counts()
```

```python
pet_food = transactions[transactions["COMMODITY_DESC"].isin( 
                         ["CAT FOOD", "DOG FOODS"])]
pet_food.info()
```

```python
pet_food["COMMODITY_DESC"].value_counts()
```

```python
#calculate regular shelf price w/o discounts
pet_food["shelf_price"] = (pet_food["SALES_VALUE"] - pet_food["RETAIL_DISC"] - pet_food["COUPON_MATCH_DISC"]) / pet_food["QUANTITY"]
#calculate paid shelf price with discounts
pet_food["paid_price"] = (pet_food["SALES_VALUE"] + pet_food["COUPON_DISC"]) / pet_food["QUANTITY"]
#calculate overall discount
pet_food["pct_disc"] = (pet_food["shelf_price"]-pet_food["paid_price"]) / pet_food["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to retailer loyalty card 
pet_food["pct_retail_disc"] = (-pet_food["RETAIL_DISC"] / pet_food["QUANTITY"]) / pet_food["shelf_price"]
#calculate discount in percentage points versus regualar shelf price due to manufacturer coupons 
pet_food["pct_coupon_disc"] = ((-pet_food["COUPON_DISC"]-pet_food["COUPON_MATCH_DISC"]) / pet_food["QUANTITY"] ) / pet_food["shelf_price"]
```

```python
#adding in causal data
causal = pd.read_parquet("data/causal_data.parquet")
causal.info()
```

```python
causal_pet_food = pd.merge(causal, products[["PRODUCT_ID", "COMMODITY_DESC"]], how = "left", on = "PRODUCT_ID")
```

```python
causal_pet_food = causal_pet_food[causal_pet_food.COMMODITY_DESC.isin(
    ["CAT FOOD", "DOG FOODS"])]
causal_pet_food.info()
```

```python
causal_pet_food["COMMODITY_DESC"].value_counts()
```

```python
pet_food = pd.merge(pet_food, causal_pet_food, how = "left", on = ["PRODUCT_ID", "STORE_ID", "WEEK_NO"])
pet_food.info()
```

```python
#remove duplicates
pet_food.drop_duplicates(subset= ["household_key", "BASKET_ID", "STORE_ID", "DAY", "PRODUCT_ID"], keep = "last", inplace= True)
pet_food.info()
```

```python
#dummy variables for display
dummies_display = pd.get_dummies(pet_food["display"], prefix = "display")
pet_food = pd.concat((pet_food, dummies_display), axis = 1)
#dummy variables for mailer
dummies_mailer = pd.get_dummies(pet_food["mailer"], prefix = "mailer")
pet_food = pd.concat((pet_food, dummies_mailer), axis = 1)
```

```python
pet_food.drop("COMMODITY_DESC_y", axis = 1, inplace = True)
pet_food.rename(columns = {"COMMODITY_DESC_x" : "COMMODITY_DESC"}, inplace=True)
pet_food.head()
```

```python
campaign_data = campaign_data[campaign_data["COUPON_UPC"] > 0]
campaign_data.info()
```

```python
pet_food = pet_food[pet_food.age_ordinal > 0]
pet_food.describe()
```

```python
#remove first 30 weeks
pet_food.drop(pet_food[pet_food.WEEK_NO < 30].index, inplace = True)
pet_food.describe()
```

```python
pet_food_wcamp = pd.merge(pet_food, campaign_data, how = "left", on = ["household_key", "PRODUCT_ID"])
pet_food_wcamp.info()
```

```python
pet_food_wcamp["CAMPAIGN_y"] = np.where((pet_food_wcamp["START_DAY"] <= pet_food_wcamp["DAY"]) & 
                                           (pet_food_wcamp["DAY"] <= pet_food_wcamp["END_DAY"]), 
                                           pet_food_wcamp["CAMPAIGN_y"], np.NaN)
pet_food_wcamp["DESCRIPTION"] = np.where((pet_food_wcamp["START_DAY"] <= pet_food_wcamp["DAY"]) & 
                                           (pet_food_wcamp["DAY"] <= pet_food_wcamp["END_DAY"]), 
                                           pet_food_wcamp["DESCRIPTION"], np.NaN)
```

```python
pet_food_wcamp["DESCRIPTION"] = np.where((pet_food_wcamp["START_DAY"] <= pet_food_wcamp["DAY"]) & 
                                           (pet_food_wcamp["DAY"] <= pet_food_wcamp["END_DAY"]), 
                                           pet_food_wcamp["DESCRIPTION"], np.NaN)
```

```python
pet_food_wcamp.drop(labels = ["MANUFACTURER_y", "DEPARTMENT_y", "BRAND_y", "COMMODITY_DESC_y", "SUB_COMMODITY_DESC_y", 
                                 "CURR_SIZE_OF_PRODUCT_y"], 
                       axis = 1, inplace = True)
```

```python
pet_food_wcamp.info()
```

```python
pet_food_wcamp.sort_values(by = ["household_key", "PRODUCT_ID", "BASKET_ID", "CAMPAIGN_y"], inplace = True)
pet_food_wcamp.drop_duplicates(subset = ["household_key", "PRODUCT_ID", "BASKET_ID"], keep = "first", inplace = True)
pet_food_wcamp.info()
```

```python
pet_food_wcamp.to_csv("data/pet_food_wcamp.csv")
```

```python
#dummify campaign and description of campaign
pet_food_wcamp = pd.read_csv("data/pet_food_wcamp.csv")
dummies_campaign_y = pd.get_dummies(pet_food_wcamp["CAMPAIGN_y"], prefix = "campaign_y")
dummies_description = pd.get_dummies(pet_food_wcamp["DESCRIPTION"], prefix = "description")
pet_food_wcamp = pd.concat((pet_food_wcamp, dummies_campaign_y), axis = 1)
pet_food_wcamp = pd.concat((pet_food_wcamp, dummies_description), axis = 1)
pet_food_wcamp.info()
```
